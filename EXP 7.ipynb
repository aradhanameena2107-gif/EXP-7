```python
# Factorial of a number

n = int(input("Enter a number: "))
fact =1

while n > 0:
  fact*= n
  n -= 1

print("Factorial =", fact)
```

```python
# Fibonacci Series using while loop

n = int(input("Enter a number: "))
a = 0
b = 1
i = 1
while i <= n:
    print(a, end =" ")
    c = a + b
    a = b
    b = c
    i += 1
```

```python
# adding limit to fibonacci series

limit = int(input("Enter limit: "))
a = 0
b = 1
while a <= limit:
  print(a, end=" ")
  a, b = b, a+b
```

```python
# Reverse a number

num = int(input("Enter a number: "))
rev = 0
while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10 # remove last digit of the number(remainder)
print("Reversed number =", rev)
```

```python
# Check Palindrome number

num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num //= 10
if temp == rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# Check Palindrome for a string
s = "madam"
i, j = 0, len(s) - 1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i += 1
  j -= 1
if is_palindrome:
  print("Yes")
else:
  print("No")
```

```python
# Taking input by the user for the string

s = str(input("Enter a string: "))
i, j = 0, len(s) - 1
is_palindrome = True
while i < j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i += 1
  j -= 1
if is_palindrome:
  print("Yes")
else:
  print("No")
```

```python
# Check Palindrome number using if else

st =input("Enter the string:")
rev = st[::-1]
if st == rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
# Count digits in a number

num = int(input("Enter a number: "))
count = 0
while num > 0:
  count += 1
  num //= 10
print("Number of digits =", count)
```

```python
# Exit the loop when i is 3:

i = 1
while i < 6:
  print(i)
  if i == 3:
   break
  i+=1
```

```python
# Search an element in a list
nums = [10, 20, 30, 40, 50]
key = int(input("Enter a number to search: "))
i = 0
while i < len(nums):
  if nums[i] == key:
    print("Element found at index", i)
    break
  i +=1
else:
    print("Element not found")
```
